```python
import redis
import json

redis_client = redis.Redis(
    host="localhost",
    port=6379,
    decode_responses=True
)
```

```python
CACHE_TTL = 60 * 60 * 6  # 6 hours

def get_jobs(query, location="India"):

    redis_key = f"jobs:query:{query.replace(' ', '_')}"
    print(redis_key)
    # 1Ô∏è‚É£ Check Redis
    cached_jobs = redis_client.get(redis_key)
    if cached_jobs:
        print("‚úÖ Redis HIT for query:", query)
        return json.loads(cached_jobs)

    print(" Redis MISS ‚Üí calling SerpAPI")

    # 2Ô∏è‚É£ Fetch from SerpAPI
    jobs = google_jobs_search(query)

    # 3Ô∏è‚É£ Store in Redis
    redis_client.setex(
        redis_key,
        CACHE_TTL,
        json.dumps(jobs)
    )

    return jobs
```

```python
from phi.agent import Agent
from phi.model.groq import Groq

# The GROQ_API_KEY is assumed to be available in the environment or global scope.

# Instantiate the Groq LLM once
groq_llm_instance = Groq(
    #id="openai/gpt-oss-120b",
    id="openai/gpt-oss-20b",
    api_key=os.environ["GROQ_API_KEY"],
    temperature=0,

    )

query_agent = Agent(
    name="Job Query Agent",
    model=groq_llm_instance,
    instructions="""
    You are a job search expert.

Your task is to convert a candidate profile into a single job search query.

STRICT RULES:
- Output ONLY the job search query text.
- Do NOT add explanations, comments, or extra words.
- Do NOT add or infer skills, tools, technologies, or frameworks.
- Use ONLY:
  ‚Ä¢ the given role
  ‚Ä¢ the given experience
  ‚Ä¢ inferred high-level fields
- Fields must be high-level domains only (e.g., software development, artificial intelligence, data science, agentic ai).
- Do NOT include specific skills such as Python, Java, React, TensorFlow, etc.
- The job search query must be meaningful and readable.

EXPERIENCE RULES:
- If experience is between 0 and 1 years, output "fresher".
- If experience is greater than 1 year, output "<N> years".

FORMAT (MUST MATCH EXACTLY):
<desired role> <experience> in <field1, field2, field3>

FIELD INFERENCE RULES:
- Infer fields ONLY from the role and context.
- Use 1‚Äì3 relevant high-level fields.
- Do NOT over-generate fields.

    """,

)


degree_agent = Agent(
    name="Degree Relevance Agent",
    model=groq_llm_instance,

    instructions="""
You are an AI job matching engine.

Your task is to evaluate the relevance between a candidate profile and a job description.

RULES:
- Evaluate degree relevance between the candidate and the job description.
- If multiple degrees are present, filter and prioritize them according to the Indian education system.
- Filter out jobs based on mismatch in Years of Experience between the candidate and the job description.
- Do NOT assume missing information.
- Do NOT add explanations outside the JSON.

DEGREE RELEVANCE RULES:
- high: Degree clearly matches job requirements
- medium: Degree partially matches job requirements
- low: Degree does not match job requirements
- If degree information is not mentioned, return "Not mentioned"

OUTPUT FORMAT:
Return STRICT JSON only with the following fields:
- degree_relevance: "high" | "medium" | "low" | "Not mentioned"
- reason: one-line explanation
- job_salary: salary if mentioned in the job description, otherwise "NA"

DO NOT return any text outside the JSON object.

""",
    temperature=0
)
```

```python
import time
def agent_2_rank_jobs(entities, jobs, top_k=10):
    resume_text = build_generic_resume_narrative(
    role=entities["p_role"],
    experience_years=entities["Years of Experience"] if "Years of Experience" in entities else "0-1",
    skills=entities["skills"]
)

    # 1Ô∏è‚É£ Rank using YOUR relevance score
    scored = []
    for job in jobs:
        score = relevance_score(resume_text, job["description"])
        scored.append({**job, "relevance_score": score[0]})

    scored.sort(key=lambda x: x["relevance_score"], reverse=True)

    # 2Ô∏è‚É£ Degree reasoning (LLM only on top-k)
    final = []
    for job in scored[:top_k]:
        prompt = f"""
Candidate Degree: {entities.get('degree', 'Not mentioned')}

Job Description:
{job['description']}
"""
        response = degree_agent.run(prompt).content

        final.append({
            "title": job["title"],
            "company": job["company"],
            "relevance_score": job["relevance_score"],
            "degree_analysis": response,
            "apply_link": job["apply_link"],
            "location": job["location"],

        })

        time.sleep(1)  # TPM safety

    return final
```

```python
def run_job_matching_pipeline(entities):
    jobs = agent_1_fetch_jobs(entities)
    final_jobs = agent_2_rank_jobs(entities, jobs)
    return final_jobs
```
